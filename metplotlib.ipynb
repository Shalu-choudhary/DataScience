<h3> Line plot</h3> <p> line plot is used to check the trend of any item and other. it is used for single column.<br> Label->> xlabel-- x axis label , ylabel- y axis label</p>

```python
ls = [3,5,7,5,8,2]
# linePLot
plt.plot(ls,marker='*',ms=15,mfc='red')   # ms = marker size , mfc = marker face color
plt.show()        # not show other extra information
```

<h3>Scatter graph</h3><p> Scatter plot is used to define relationship between two axis . <br>in this two columns are used .it is used when both values are numerical</p>
 <h4> corelation</h4><p> <h5>+ve relation</h5> when both x and y are increase simutaneously <h5> -ve relation</h5> is when one is increase and another is decrease both are inverse to each other

```python
price=list(df['price'].values)
age=list(df['age'].values)
plt.scatter(age,price,color='red')
plt.xlabel('price')
plt.ylabel('age')
plt.title('AGE V/s SHOW')
plt.show()
```

<h4>Bar Graph </h4>
<p> bar graph is used when one value is catagorical and other one is numerical <br> 1)  ec-->> edge color  2)  lw--->> line width  3)  ls--> line style . Line style is solid bydefault </p> <br> <p> Figsize->>>> it is used for remove overlapping or increase the width and height of the graph size .<br> in this two parameter is used (width,height)</pr>

```python
brand=list(df['brand'].value_counts().keys())        # CATAGORICAL
brand_count=list(df['brand'].value_counts().values)    #NUMERICAL
```

```python
brand
```

```python
plt.bar(brand,brand_count,color='green',ec='red',ls="solid",lw=2,width=0.8)
# plt.figure(figsize=(36,8))
plt.xlabel("barnds name")
plt.ylabel(" brands no. ")
plt.title("no of bikes in brand")
plt.xticks(rotation=90)
plt.show()
```

```python
arr=np.random.randint(1,100,50)
bin=[0,20,50,60,75,90]
plt.xlabel('age')
plt.ylabel("people")
plt.title('Age Chart')
plt.hist(arr,color='pink',bins=bin,ec='blue',lw=2)
plt.show()
```
