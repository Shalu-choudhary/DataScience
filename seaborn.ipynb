```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.columns=['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'member']
```

```python
df
```

catplot---> catgorical plot is used when one column is catgorical and one columns is numerical<br>

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker')
plt.show()
#hue---> add extra info(col)
```

```python
#col--->> for seprating
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='gender')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',col='day')
plt.show()
```

```python
female=df[df['gender']=='Female']
```

```python
sns.relplot(x='tip',y='total_bill',data=female)
```

```python
sns.relplot(x='total_bill',y='tip',data=female,hue='smoker',col='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=female,hue='smoker',row='day')
plt.show()
```

```python
sns.relplot(x='survived',y='age',data=df2)
plt.show()
```

```python
sns.relplot(x='survived',y='age',data=df2,hue='gender',style='male/female',col='male/female')
plt.show()
```

```python
sns.relplot(x='survived',y='age',data=df2,hue='gender',style='male/female',col='male/female',kind='line')
plt.show()
```

```python
sns.relplot(x='survived',y='age',data=female,hue='gender',col='male/female')
plt.show()
```
